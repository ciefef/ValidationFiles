# Converting CSV to KML with Multiple icon

After removing duplicate boats in excel, this file converts the csv to kml with different icons based on the vessel type of the boats. The output of this jupyter notebook is kml file.

```python
# Importing CSV files
import pandas as pd
import  simplekml

# Importing csv files
newfile = 'Oca_Modified_AIS-MultipleVT_Removed_duplicates.csv'
ais_new = pd.read_csv(newfile)

print (ais_new)
```
